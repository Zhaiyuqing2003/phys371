```python
import numpy as np
import matplotlib.pyplot as plt

# Read lines from the file
lines = open("DATA.TXT", "r").read().splitlines()

times = []
xs = []
ys = []
zs = []

for l in lines:
    entries = l.split(' ')
    if len(entries) == 5:
        times.append(int(entries[0])/1000_000)
        xs.append(float(entries[1]))
        ys.append(float(entries[2]))
        zs.append(float(entries[3]))

# Use this to specify the needed range.
range_select = [0, -1, 1]

def plot(ns, nlabel):
    plt.plot(times[range_select[0]:range_select[1]:range_select[2]], ns[range_select[0]:range_select[1]:range_select[2]])
    plt.xlabel("Time (s)")
    plt.ylabel(nlabel + " (m/s^2)")
    plt.show()
    
# Example
plot(xs, 'X')
plot(ys, 'Y')
plot(zs, 'Z')
```
